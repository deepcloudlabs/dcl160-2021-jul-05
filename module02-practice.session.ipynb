```python
departments = set([])
for full_name, department, *rest in employees.values():
  departments.add(department)
departments = sorted(departments)
print(departments)  
```

```python
emp_numbers_by_dept = {} # key: department name (str) , value: number of employees in that department (int)
for department in departments:
  emp_numbers_by_dept[department] = 0
print(emp_numbers_by_dept)  
for full_name, department, *rest in employees.values():
  emp_numbers_by_dept[department] += 1
print(emp_numbers_by_dept)  
```

```python
emp_numbers_by_dept = {}
for full_name, department, *rest in employees.values():
  if department not in emp_numbers_by_dept:
    emp_numbers_by_dept[department] = 1
  else:
    emp_numbers_by_dept[department] += 1  
print(emp_numbers_by_dept)    
```
