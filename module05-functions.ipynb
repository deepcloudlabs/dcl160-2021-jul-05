```python
# classical function
def _3n_plus_one(n):
  sequence = [n]
  while n > 1:
    n = n//2 if n%2 == 0 else 3*n+1
    sequence.append(n)
    print(f"_3n_plus_one() is adding the value {n} to the list")
  return sequence  
```

```python
# generator function -> lazy -> iterable
def gen_3n_plus_one(n):
  print(f"gen_3n_plus_one({n}) is just started!")
  yield n
  while n > 1:
    n = n//2 if n%2 == 0 else 3*n+1
    print(f"gen_3n_plus_one() is yielding the value {n}")
    yield n
```

```python
result = gen_3n_plus_one(17)
```

```python
next(result)
```

```python
for number in gen_3n_plus_one(17): # no memory consumption!
  print(number)
```
