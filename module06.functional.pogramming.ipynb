```python
numbers = (4,8,15,16,23,42)
```

```python
#problem: take the sum of square of even numbers
```

```python
#procedural programming -> structured programming
def is_even(num):
  return num%2 == 0

def take_square(number):
  return number ** 2

def solve_problem(*values):
  solution = 0
  for value in values:
    if is_even(value):
      squared = take_square(value)
      solution += squared
  return solution    
```

```python
solve_problem(4,8,15,16,23,42)
```

```python
#functional programming -> filter/map/reduce
#                          filter -> generator function -> lazy  -> yield
#                          map    -> generator function -> lazy  -> yield
#                          sum    -> classical function -> eager -> return
result = sum(map(take_square , filter(is_even , numbers ))) # function pipeline -> generator
#filter : function -> is_even -> function
```

```python
print(result)
```

```python
from functools import reduce
```

```python
def my_sum(acc,val):
  return acc+val
```

```python
# reduce -> eager function -> returns the result
result = reduce( my_sum , map(take_square , filter(is_even , numbers )) , 0 )
```

```python
print(result)
```

```python
# functional programming -> def / lambda expression -> function -> unnamed
```

```python
reduce( lambda acc,val : acc+val , map(lambda number : number ** 2 , filter(lambda num : num%2 == 0, numbers )) , 0 )
```

```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,False),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1979,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 170000, 1984,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]  
```

```python
from functools import reduce
# find oldest employee
def min_birth_year_reducer( min_emp , curr_emp):
  if min_emp is None:
    return curr_emp
  return min_emp if min_emp[3] < curr_emp[3] else curr_emp
reduce( min_birth_year_reducer  , employees , None )
```

```python
# find total/min/max/avg salary
stats = {
    "min": float("inf"),
    "max": -float("inf"),
    "avg": 0,
    "total": 0,
    "count": 0 
}
def salary_statistics_reducer(statistics , salary):
    stats["count"] += 1
    stats["total"] += salary
    stats["min"] = stats["min"] if stats["min"] < salary else salary
    stats["max"] = stats["max"] if stats["max"] > salary else salary  
    stats["avg"] = stats["total"] / stats["count"]   
    return statistics

reduce(salary_statistics_reducer , map(lambda emp : emp[2] , employees) ,stats )
```
